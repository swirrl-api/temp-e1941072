```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import matplotlib.cm as cm
```

```python
filename = '~/work/private/ECMWF_2022_02_28.nc'
```

```python
data = xr.open_dataset(filename)
```

```python
V = np.linspace(-20, 20, 30)
```

```python
fig = plt.figure(figsize=[12,8])
proj = ccrs.PlateCarree()

ax = plt.axes(projection=proj)

CS = ax.contourf(data.longitude, data.latitude, data.u10[2,:,:], V, cmap = cm.seismic, alpha = 0.35, transform = ccrs.PlateCarree())  #divide by 10 is conversion to decameters

ax.coastlines(resolution='10m', color='blue', linewidth=2)
ax.gridlines()
plt.show()
#V = np.linspace(300,650,30)
```
